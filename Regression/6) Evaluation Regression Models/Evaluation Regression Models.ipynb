# Evaluate Regression Models

**Goal:** Quantify how well a regression model fits the data using standard metrics.

| Metric | Formula | Meaning |
|--------|---------|---------|
| SSR | Σ(yᵢ - ŷᵢ)² | Error *not* explained by the model |
| SST | Σ(yᵢ - ȳ)² | Total variation in the data |
| R² | 1 − SSR/SST | Fraction of variance explained (1 = perfect, 0 = no better than mean) |

### import libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### import data

```python
df = pd.read_csv("linear-regression-dataset.csv",sep=";")
df
```

### preparing the data to fit

The same salary/experience dataset from the Linear Regression notebook is reused here. `x` = experience, `y` = salary.

```python
x = df.experience.values.reshape(-1,1)
y = df.salary.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(x,y)
y_head = lg.predict(x)
```

### plotting the results

Green dots = actual data, red line = model predictions.  
The closer the line is to the dots, the better the fit.

```python
plt.scatter(x,y,color="green")
plt.plot(x,y_head,color="red")
plt.show()
```

### model r2_score

`r2_score` from sklearn computes R² directly.  
A value close to **1.0** means the model explains almost all variance in the data — great fit! Values below 0 indicate the model is worse than simply predicting the mean.

```python
from sklearn.metrics import r2_score
print(r2_score(y,y_head))
```
