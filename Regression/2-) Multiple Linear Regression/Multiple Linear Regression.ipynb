# Multiple Linear Regression

**Goal:** Extend linear regression to multiple input features.
Here we predict `salary` from both `experience` and `age (yas)`.
Formula: `ŷ = b0 + b1·x1 + b2·x2`

### import libraries

```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
```

### import data

```python
df = pd.read_csv("multiple-linear-regression-dataset.csv",sep=";")
df.head()
```

###  preparing the x and y for predict

`iloc[:,[0,2]]` selects columns 0 (experience) and 2 (age) as the feature matrix X.  
`reshape(-1,1)` turns the salary series into a column vector required by sklearn.

```python
x = df.iloc[:,[0,2]].values
y = df.salary.values.reshape(-1,1)
```

### linear regression

```python
multiple_linear_regression = LinearRegression()
multiple_linear_regression.fit(x,y)
```

```python
print("b0: ",multiple_linear_regression.intercept_)
print("b1,b2: ",multiple_linear_regression.coef_)
```

### predict

Predicting salary for three hypothetical employees:
- 5 years experience, age 35
- 10 years experience, age 35
- 20 years experience, age 30

Notice how age has a *negative* coefficient (`b2 < 0`): for the same experience level, an older employee actually earns slightly less in this small dataset — a quirk of the data, not a general rule.

```python
multiple_linear_regression.predict(np.array([[5,35],[10,35],[20,30]])) # x= experience, y = age
```
